```python
import easyocr
import cv2
from google.colab.patches import cv2_imshow
```

```python
reader = easyocr.Reader(['en'])
```

```python
result = reader.readtext('/content/eng 3.jpg' ,paragraph=True)
```

```python
text = ' '.join([res[1] for res in result])
```

```python
print(text)
```
